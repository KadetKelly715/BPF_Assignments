```python
# Sales Data Values
# 2022 Data
x = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec']
y = [8000, 6000, 9000, 14000, 11000, 10500, 9500, 7000, 9000, 8500, 11000, 9000]
plt.plot(x,y, color='green', linestyle='--', label = '2022 Sales by Month')

# 2023 Data
x2 = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec']
y2 = np.random.randint(6000, 14000, 12)
plt.plot(x2, y2, color='grey', linestyle=':', label = '2023 Sales by Month')

plt.title('Sales by Month')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()

plt.show()
```

```python
dfFord = pd.read_csv('drive/MyDrive/F.csv')
dfTesla = pd.read_csv('drive/MyDrive/TSLA.csv')

x_Ford = dfFord['Date']
y_Ford = dfFord['Close']

plt.plot(x_Ford, y_Ford, color = 'orange', linestyle = '--', label = 'Ford 2023 Closing')

x_Tesla = dfTesla['Date']
y_Tesla = dfTesla['Close']

plt.plot(x_Tesla, y_Tesla, color = 'purple', label = 'Tesla 2023 Closing')

plt.title('Ford vs. Tesla')
plt.xlabel('Date')
plt.ylabel('Closing Stock Price ($)')
plt.legend()
plt.show()
```
